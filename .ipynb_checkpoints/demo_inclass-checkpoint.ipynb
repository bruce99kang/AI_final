```python
import os
import sys
import cv2
import matplotlib.pyplot as plt
import numpy as np 
from imutils import paths

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model

import math
import pandas as pd
from tensorflow.keras import preprocessing
from model.model import *

model = tf.keras.models.load_model('./work_dir/my_model_6class_blanced.h5')
data_test = pd.read_csv('./data_dataframe/total_dataframe.csv',header=0)
to_train = math.floor(0.8*len(data_test))
to_val = math.floor(0.9*len(data_test))
choose = ['Bangs','Big_Nose','Bushy_Eyebrows','Chubby','Eyeglasses','Mustache']
test_datagen = preprocessing.image.ImageDataGenerator(rescale=1./255)


testing_generator = test_datagen.flow_from_dataframe(
    dataframe=data_test.iloc[to_val:],
    directory='/home/bruce/Downloads/CelebA/CelebA-20201202T065512Z-015/CelebA/Img/img_align_celeba/',
    x_col='images',
    y_col=choose,
    target_size=(300,300),
    class_mode='other',
    batch_size=64)
```

```python
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt

testing_images = data_test.iloc[to_val:]
for i in range(to_val+100, to_val+130,2):
    img_ori = cv2.imread('/home/bruce/Downloads/CelebA/CelebA-20201202T065512Z-015/CelebA/Img/img_align_celeba/'+testing_images.images[i])
    image = cv2.resize(img_ori, (300, 300))
    imagearray = img_to_array(image)

    data=np.expand_dims(imagearray, axis=0)
    result=model.predict(data)
    print(result)
    print(type(result))
    r=np.argmax(result)
    
    fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2)
    ax1.imshow(img_ori)
    if np.sum(result)==0:
        ax1.set_title('Predict '+"None")
    else:
        ax1.set_title('Predict '+choose[r])
    ax2.imshow(image)
    ax2.set_title('resized')
    plt.show()
```
